```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import os
import datetime as dt
```

```python
nat_disasters_path = "Natural disasters.csv"
nat_disasters_cost_path = "disaster_cost.csv"
nat_disastersdata = pd.read_csv(nat_disasters_path)
nat_disastercost = pd.read_csv(nat_disasters_cost_path)
# nat_disastersdata
# nat_disastercost

frames = [nat_disastersdata, nat_disastercost]

result = pd.concat(frames)
result
```

```python
result["Year"]= result["Year"].astype(str)
```

```python
nat_disaster_year = result.sort_values(by = ['Year'],ascending = False)
nat_disaster_year.reset_index()
```

```python
nat_dis_21 = nat_disaster_year.drop(nat_disaster_year[nat_disaster_year.Year < "2000"].index)
nat_dis_21
```

```python
nd_txca = nat_dis_21.loc[(nat_dis_21['state'] == "CA") | (nat_dis_21['state'] == "TX" )]
nd_txca
nd_txca.to_csv('TX_CA_Disasters.csv')
```

```python
tx_ca_disasters = "TX_CA_Disasters.csv"
tx_ca_disasters_df = pd.read_csv(tx_ca_disasters)
tx_ca_disasters_df
```
