```python
# Dependencies and Setup
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import scipy.stats as st
import numpy as np
import os
import datetime as dt
```

```python
nat_disasters_path = "Natural disasters.csv"
nat_disasters_cost_path = "disaster_cost.csv"
nat_disastersdata = pd.read_csv(nat_disasters_path)
nat_disastercost = pd.read_csv(nat_disasters_cost_path)
```

```python
nd_ca = nat_disastercost.loc[(nat_disastercost['state'] == "CA")]
nd_ca
```

```python
curr_nat_dis_cost = nd_ca.drop(nd_ca[nd_ca.Year < "2000"].index)
curr_nat_dis_cost
```

```python
curr_nat_dis_cost["Damage cost (billions USD)"]=curr_nat_dis_cost["Damage cost (billions USD)"].astype(float)
```

```python
yearly_nat_disastercost_ca = curr_nat_dis_cost.groupby('Year')['Damage cost (billions USD)'].sum()
yearly_nat_disastercost_ca = pd.DataFrame(data=yearly_nat_disastercost_ca)
yearly_nat_disastercost_ca
```

```python
# x_line = yearly_nat_disastercost_ca["Year"]
# y_line = yearly_nat_disastercost_ca["Damage cost (billions USD)"]

# line_graph, = plt.plot(x_line, y_line, color="blue")
# plt.title("Natural Disaster Cost per Year in California")
# plt.xlabel("Year")
# plt.ylabel("Damage cost (billions USD)")
# plt.legend(handles=[line_graph], loc="best")
# plt.show()

yearly_nat_disastercost_ca.plot(kind="bar")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.ylabel("Damage Cost (billions USD)")
plt.title("Natural Disaster Cost per Year in California")
plt.legend("")
plt.show()
```

```python
nat_disastersdata["Year"]=nat_disastersdata["Year"].astype(str)
nat_disastersdata["Year"].describe()
```

```python
nat_dis_ca = nat_disastersdata.loc[(nat_disastersdata['state'] == "CA")]
nat_dis_ca
```

```python
curr_nat_dis_ca = nat_dis_ca.drop(nat_dis_ca[nat_dis_ca.Year < "2000"].index)
# curr_nat_dis_ca

yearly_nat_dis_cal = curr_nat_dis_ca.groupby("Year")
curr_nat_dis_ca_total = yearly_nat_dis_cal["Year"].count()
# curr_nat_dis_ca_total

curr_nat_dis_ca_total_df = pd.DataFrame(curr_nat_dis_ca_total)
curr_nat_dis_ca_total_df = curr_nat_dis_ca_total_df.rename(columns={"Year": "Number of Natural Disasters"})
curr_nat_dis_ca_total_df
```

```python
curr_nat_dis_ca_total_df = curr_nat_dis_ca_total_df.sort_values(by = ['Year'],ascending = True)
curr_nat_dis_ca_total_df = curr_nat_dis_ca_total_df.reset_index()
curr_nat_dis_ca_total_df


# time = curr_nat_dis_ca_total_df["Year"]
# disasters = curr_nat_dis_ca_total_df["Number of Natural Disasters"]
# plt.plot(time, disasters)
# plt.show()
```

```python
x_line = curr_nat_dis_ca_total_df["Year"]
y_line = curr_nat_dis_ca_total_df["Number of Natural Disasters"]

plt.figure(figsize = (12,8))
line_graph, = plt.plot(x_line, y_line, color="blue", label="California")
plt.title("Total Number of Natural Disasters in California by Year")
plt.xlabel("Year")
plt.ylabel("Number of Natural Disasters")
plt.legend(handles=[line_graph], loc="best")
plt.xticks(rotation=45)
plt.show()

# time = curr_nat_dis_ca_total_df["Year"]
# disasters = curr_nat_dis_ca_total_df["Number of Natural Disasters"]
# plt.figure(figsize=(12,8))
# plt.plot(time, disasters)
# plt.title("Total Number of Natural Disasters in California by Year")
# plt.xlabel("Year")
# plt.ylabel("Number of Natural Disasters")
# plt.legend(handles=[line_graph], loc="best")
# plt.xticks(rotation=45)
# plt.show()
```
