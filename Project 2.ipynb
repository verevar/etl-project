```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import os
import datetime as dt
```

```python
nat_disasters_path = "Natural disasters.csv"
nat_disasters_cost_path = "disaster_cost.csv"
nat_disastersdata = pd.read_csv(nat_disasters_path)
nat_disastercost = pd.read_csv(nat_disasters_cost_path)
# nat_disastersdata
# nat_disastercost

frames = [nat_disastersdata, nat_disastercost]

result = pd.concat(frames)
result
```

```python
result["Year"]= result["Year"].astype(str)
```

```python
nat_disaster_year = result.sort_values(by = ['Year'])
nat_disaster_year.reset_index()
```

```python
nat_disaster_type = nat_disaster_year.groupby("incident_type")
nat_disaster_type = nat_disaster_type["incident_type"].count()
nat_disaster_type
```
